```python
def to_any(num, base):
    r = ''
    
    while num > 0:
        r += str(num % base)
        num //= base
    
    return r


def algo(n):
    r = to_any(n, 3)
    if sum([int(i) for i in r]) % 2 == 0:
        r += '0'
        r = '2' + r[2:]
    else:
        r += '1'
        r = '20' + r[2:]
    
    return int(r, 3)


rns = []

for n in range(1, 1000):
    if algo(n) > 75:
        rns.append([algo(n), n])
        

sorted(rns, reverse=True)
```

```python
def dels(num):
    ds = []
    
    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            ds.append(i)
            ds.append(num//i)
    
    if ds == []:
        return 0
    else:
        return sum(ds)
    
    
for num in range(424242, 10**15):
    if dels(num) % 2024 == 42:
        print(dels(num))

```
