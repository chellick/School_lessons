```python
def algo(n):
    r = bin(n)[2:] 
    if n % 3 == 0:
        r += r[-2] + r[-1]
    else:
        r += bin(n%3*3)[2:]
        
    return int(r, 2)

rs = []

for n in range(3, 1000):
    if algo(n) >= 195:
        rs.append(algo(n))
    
min(rs)
```

```python
with open('9.txt') as f:
    nums = list(list(map(int, i.split())) for i in f.readlines())
    
def check(row):
    if len(set(row)) == 6:
        row = sorted(row)
        
        nreps = []
        reps = []
        
        for i in row:
            if row.count(i) == 2:
                reps.append(i)
            else:
                nreps.append(i)
                
        total=1
        
        for i in range(3):
            total *= nreps[i]
        
        if total > reps[0] ** 2:
            return True
    return False


c = 0
for row in nums:
    if check(row):
        c += 1
        
c
```

```python
sums = []

for t in range(4, 1000):
    s = '8' + '4' * t

    while '11' in s or '444' in s or '8888' in s:
        if '11' in s:
            s = s.replace('11', '4', 1)
        if '444' in s:
            s = s.replace('444', '88', 1)
        if '8888' in s:
            s = s.replace('8888', '1', 1)

    sums.append(sum([int(i) for i in s]))

max(sums)
```

```python
with open('17.txt') as f:
    maxelem = 0
    nums = []
    
    for i in f.readlines():
        i = int(i)
        if i % 100 == 21:
            if 9999 < abs(i) < 100000:
                maxelem = max(maxelem, i) 
        nums.append(i)


def pairch(pair):
    st21 = []
    
    for i in pair:
        if (abs(i) % 100 == 21) and (9999 < abs(i) < 100000):
            st21.append(i)
    
    if len(st21) == 1:
        if (pair[0] ** 2 + pair[1] ** 2) >= (maxelem ** 2):
            return True
    return False
    

c = []
for i in range(len(nums)-1):
    if pairch([nums[i], nums[i+1]]):
        c.append(sum([nums[i], nums[i+1]]))
        
len(c), max(c)    
```

```python
with open('24.txt') as f:
    letters = f.read()
    letters = letters.replace('KLMN', '-')
    # letters = 'LMN--KLMMMMMMM'

state = False

curlen = 0
maxlen = []

for i in range(len(letters)):
    try:
        if letters[i] == '-':
            curlen += 4
            if letters[i+1] == '-':
                state = True
            else:
                state = False
        
        if state:
            if letters[i-1] == 'N':
                curlen += 1
                if letters[i-2] == 'M':
                    curlen += 1
                    if letters[i-3] == 'L':
                        curlen += 1
            else:
                continue
        
        if not state:
            if letters[i+1] == 'K':
                curlen += 1
                if letters[i+2] == 'L':
                    curlen += 1
                    if letters[i+3] == 'M':
                        curlen += 1
            
            maxlen.append(curlen)
            curlen = 0
            # print(max(maxlen))
    except:
        pass
        
            
        
max(maxlen)
        

```
