```python
with open('pro100ege.txt') as f:
    nums = list(list(map(int, i.split())) for i in f.readlines())
    
def sub_process(row):
    r = [max(row[:2]), row[-1]]
    return r

nums = [sub_process(i) for i in nums]
nums = sorted(nums)
nums

for sec in range(0, 100):
    current = 0
    
    for proc in nums:
        if proc[0] <= sec <= proc[1]:
            current += 1
    
    print(current, sec)
```
