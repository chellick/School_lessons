# MySQL Database

SQL (Structured Query Language) - это язык программирования, который используется для работы с реляционными базами данных. SQL позволяет создавать, изменять, удалять и извлекать данные из базы данных. Он используется в различных областях, таких как банковское дело, медицинские учреждения, интернет-магазины и многие другие.

MySQL - это система управления реляционными базами данных, которая использует SQL для управления данными. MySQL является одним из наиболее популярных серверов баз данных в мире и широко используется веб-разработчиками. MySQL является бесплатным и открытым исходным кодом, что означает, что его можно свободно загружать, использовать и изменять.

MySQL позволяет создавать базы данных, таблицы и связи между ними, а также добавлять, изменять и удалять данные. Он также поддерживает многопользовательский доступ и безопасность для защиты ваших данных. MySQL можно использовать с различными языками программирования, такими как PHP, Python, Java и другие.

В целом, SQL и MySQL являются важными инструментами для хранения, управления и извлечения данных, и их использование в современном мире является необходимым для различных видов приложений и сервисов.

````python
# Подключение к MySQL
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="qwer1324",
  database="school"
)
mycursor = mydb.cursor()
````

````python

# создаем базу данных
mycursor.execute("CREATE DATABASE school")
# Выбираем базу данных 
mycursor.execute("USE school")
# Создаём таблицу
mycursor.execute("CREATE TABLE students (id INT AUTO_INCREMENT PRIMARY KEY, name VARCHAR(255) NOT NULL, age INT)")
````

````python
# Добавление значений в таблицу
mycursor.execute("INSERT into students VALUES(2, 'Matvey Isupov', 16)")
````

````python
# Выполнение запроса на выборку данных из таблицы
mycursor.execute("SELECT * FROM students")
# Получение результатов запроса
myresult = mycursor.fetchall()
# Вывод результатов на экран
for row in myresult:
  print(row)
````
