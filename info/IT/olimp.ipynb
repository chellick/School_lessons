# Олимпиадное программирование

Олимпиадное программирование представляет собой специфическую область программирования, в которой необходимо обладать не только глубокими знаниями языка программирования, но и уметь решать сложные алгоритмические задачи.

Вот несколько тем, которые необходимо знать для олимпиадного программирования:

Алгоритмы сортировки: пузырьковая, вставками, выбором, быстрая, слиянием и другие.

Поиск элементов: бинарный поиск, поиск в глубину (DFS), поиск в ширину (BFS).

Графы: обход графов, поиск кратчайшего пути, топологическая сортировка.

Динамическое программирование: поиск наибольшей общей подпоследовательности, рюкзак, редакционное расстояние и другие.

Математические алгоритмы: нахождение простых чисел, поиск наибольшего общего делителя (НОД), нахождение наименьшего общего кратного (НОК), преобразование Фурье и другие.

Структуры данных: стек, очередь, деревья, хеш-таблицы, кучи, сбалансированные деревья поиска и другие.

Теория графов: связность, циклы, деревья, максимальный поток и другие.

Параллельное программирование: работа с потоками, процессами, синхронизация.

Игры и задачи на логику: игры с ними, задачи на поиск решений и другие.
