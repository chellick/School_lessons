Уравенние прямой проведённой через две заданные точки


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


x1 = 2
y1 = 4
x2 = 6
y2 = 1

x = np.linspace(3, 6, 100)

def equation(x, x1, y1, x2, y2):
    if x2 - x1 == 0:
        x = np.linspace(x1, x2 + 0.000001, 100)
        x2 += 0.000001
        return ((x - x1) * (y2 - y1) + y1 * (x2 - x1)) / (x2 - x1)
    else:  
        return ((x - x1) * (y2 - y1) + y1 * (x2 - x1)) / (x2 - x1)


y = equation(x, x1, y1, x2, y2)
y_one = np.linspace(6, 6, 100)
# print(y_one)

plot = plt.scatter(x, y)
plot = plt.scatter([x1, x2], [y1, y2])

plt.show()


```

```python
x1 = 3
y1 = 7
x2 = 6
y2 = 3
x3 = 6
y3 = 7

line1 = equation(x, x1, y1, x2, y2)
line2 = equation(y_one, x2, y2, x3, y3)
line3 = equation(x, x1, y1, x3, y3)


plot = plt.scatter(x, line1)
plot = plt.scatter(y_one, line2)
plot = plt.scatter(x, line3)

plt.show()


```
